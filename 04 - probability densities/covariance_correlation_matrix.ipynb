**COVARIANCE MATRIX SANITY CHECK**

```python
import numpy as np
import matplotlib.pyplot as plt

Nobs = 1000

ms = [
    np.random.multivariate_normal(
        np.array([-1.5, 3.4]), 
        np.array([
            [1.6, 1.3],
            [1.3, 1.2]
        ]), 
        Nobs * 5//11
    ).T,
    np.random.multivariate_normal(
        np.array([10.1, -7.2]), 
        np.array([
            [2.4, 1.6],
            [1.6, 3.0]
        ]), 
        Nobs * 1//11
    ).T,
    np.random.multivariate_normal(
        np.array([9.9, -8.8]), 
        np.array([
            [1.6 ,-1.7],
            [-1.7, 2.9]
        ]), 
        Nobs * 5//11
    ).T,
]


# plt.gca().set_aspect('equal')
# plt.xlim([-40, 40])
# plt.ylim([-40, 40])
for i, m in enumerate(ms):
    plt.scatter(m[0], m[1], label=i)
plt.legend()
```
