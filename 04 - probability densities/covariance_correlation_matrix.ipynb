**COVARIANCE MATRIX SANITY CHECK**

```python
import numpy as np
import matplotlib.pyplot as plt

Nobs = 1000

ms = [
    np.random.multivariate_normal(
        np.array([-0.5, -4.6]), 
        np.array([
            [1.7, -1.3],
            [-1.3, 2.1]
        ]), 
        Nobs * 1//2
    ).T,    
    np.random.multivariate_normal(
        np.array([-5.8, 13.1]), 
        np.array([
            [2.1, -1.6],
            [-1.6, 2.4]

        ]), 
        Nobs * 1//10
    ).T,    
    np.random.multivariate_normal(
        np.array([2.5, 1]), 
        np.array([
            [2.7, 1],
            [1, 1.4]
        ]), 
        Nobs * 2//5
    ).T,    
]


plt.gca().set_aspect('equal')
plt.xlim([-10, 10])
plt.ylim([-10, 20])
for i, m in enumerate(ms):
    plt.scatter(m[0], m[1], label=i)
plt.legend()
```
