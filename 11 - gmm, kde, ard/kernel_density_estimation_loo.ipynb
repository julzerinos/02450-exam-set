**LOO CURVE**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

xs = np.array([3.918, -6.35, -2.677, -3.003])
N = len(xs)

sigmas = np.linspace(2, 8, 50)


def E(sigma):
    E = 0
    for n in range(N):
        xn = xs[n]

        S = 0
        for i in range(N):
            if i == n: continue
            S += norm.pdf(xn, xs[i], sigma)
        
        E += np.log(1/(N-1) * S)

    return -1/4 * E

plt.plot(sigmas, E(sigmas))
```
