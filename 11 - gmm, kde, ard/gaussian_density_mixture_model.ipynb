**CALCULATE GAUSSIAN DENSITY**

```python
import numpy as np

distances = np.array([2.11, 1.15, 1.09])
K = distances.size
sigma = .5
dimensionality = 10
SIGMA = np.array([
    [1, 0], 
    [0, 1]
])



gaussian = 1/(2 * np.pi * sigma**2 * np.linalg.det(SIGMA))**(dimensionality/2) * np.e**(-distances**2/(2 * sigma**2))
density = np.sum(gaussian) / K

density
```

**EM STEP WITH GRAPH PROBABILITIES**

```python
import numpy as np
from scipy.stats import norm


weights = np.array([.19, .34, .48])
probabilities = np.array([
    norm.pdf(3.19, 3.177, .0062), 
    norm.pdf(3.19, 3.181, .0076), 
    norm.pdf(3.19, 3.184, .0075), 
])

for i, p in enumerate(probabilities):
    pw = p * weights[i]
    others = (probabilities[probabilities != p] * weights[probabilities != p]).sum()
    posterior = pw / (pw + others)
    print(f"posterior for point belonging to component {i}: {posterior}")
```
