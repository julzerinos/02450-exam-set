**PURITY GAIN, IMPURITY**


```python
import numpy as np

gini = lambda a: 1 - ((a/a.sum())**2).sum()
class_err = lambda a: 1 - np.max(a/a.sum())
# entropy = lambda c1, c2: 

impurity = class_err

root = np.array([1, 2, 3, 4])

branches = [
    np.array([143, 137, 54, 1]),
    np.array([0, 2, 3, 5]),
]

branches_impurity_gain = 0
for i, b in enumerate(branches):
    branch_impurity_gain = (b.sum()/root.sum()) * impurity(b)
    branches_impurity_gain += branch_impurity_gain
    print(f"impurity gain for branch {i}: {branch_impurity_gain}")

purity = impurity(root) - branches_impurity_gain

print()
print("impurity of the root:", impurity(root))
print("purity gain of the split:", purity)
```

**HUNT'S REGRESSION TREE**

```python
import numpy as np

root = np.array([23, 6, 13, 14, 17])

branches = [
    np.array([23]),
    np.array([6, 13, 14, 17]),
]

impurity = lambda n: np.mean( (n - np.mean(n))**2 )

branches_impurity_gain = 0
for i, b in enumerate(branches):
    branch_impurity_gain = (b.size/root.size) * impurity(b)
    branches_impurity_gain += branch_impurity_gain
    print(f"impurity gain for branch {i}: {branch_impurity_gain}")

purity = impurity(root) - branches_impurity_gain

print()
print("impurity of the root:", impurity(root))
print("purity gain of the split:", purity)
```
