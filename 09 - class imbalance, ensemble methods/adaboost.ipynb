**ADABOOST, FIRST STEP**

```python
import numpy as np

incorrect = np.array([0, 1, 1, 1, 0, 1, 1]) 

w = np.ones(len(incorrect))/len(incorrect)
err = (incorrect * w).sum()
alpha = .5 * np.log((1-err)/err)

incorrect_weight = w[0] * np.e ** (alpha)
correct_weight = w[0] * np.e ** (-alpha)

print(f"new incorrect weight: {incorrect_weight / (incorrect.sum() * incorrect_weight + (len(incorrect) - incorrect.sum()) * correct_weight)}")
print(f"new correct weight: {correct_weight / (incorrect.sum() * incorrect_weight + (len(incorrect) - incorrect.sum()) * correct_weight)}")
```

**COMBINING T ADABOOST CLASSIFIERS**


```python
import numpy as np

classes = [1, 2]
ys = [
    np.array([2, 1, 1, 2]),
    np.array([2, 2, 1, 2])
]

alpha = np.array([.583, .657, .591, .398])

for i, y in enumerate(ys):

    alpha_tests = []
    for c in classes:
        alpha_tests.append(np.sum(alpha[y == c]))

    print(f"adaboost declares class for y{i}: << {classes[np.argmin(alpha_tests)]} >>")
```
