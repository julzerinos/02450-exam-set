**PRECISION, RECALL, TPR, FRP**

```python
TP = 2  # TPR, precision
TN = 2  # FPR
FP = 3  # precision, FPR
FN = 2 # TPR

print("sum of points", TP+TN+FP+FN)

print(f"precision: \t{TP/(TP+FP)}")
print(f"recall/TPR: \t{TP/(TP+FN)}")
print(f"FPR: \t\t{FP/(TN+FP)}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics

data = np.array([0, .05, .1, .15, .38, .5, .54, .72, .84])
data_class = np.array([1,1,1,0,0,0,1,0,1])

fpr, tpr, thresholds = metrics.roc_curve(data_class, data)
plt.plot(fpr, tpr, 'r', [0, 1], [0, 1], 'k')
```
